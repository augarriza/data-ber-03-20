##  Imports

```python
import pandas as pd
import requests
import json
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

# Bitcoin market value for last 30 days, vs. USD, deliver a DF and a Plot

```python
res_bit = requests.get("https://api.coingecko.com/api/v3/coins/bitcoin/market_chart?vs_currency=usd&days=30")
j_bit = res_bit.json()
df_bit = pd.DataFrame(j_bit["prices"], columns = ["time", "price"])
df_bit["time"] = pd.to_datetime(df_bit["time"], unit='ms')
df_bit.set_index('time', inplace=True)
df_bit.plot()
```

# Litecoin market value for last 30 days, vs. USD, deliver a DF and a Plot

```python
res_lit = requests.get("https://api.coingecko.com/api/v3/coins/litecoin/market_chart?vs_currency=usd&days=30")
j_lit = res_lit.json()
df_lit = pd.DataFrame(j_lit["prices"], columns = ["time", "price"])
df_lit["time"] = pd.to_datetime(df_lit["time"], unit='ms')
df_lit.set_index('time', inplace=True)
df_lit.plot()
```

## Plot Litecoin and Dogecoin prices in a lineplot for each

```python
res_dog = requests.get("https://api.coingecko.com/api/v3/coins/dogecoin/market_chart?vs_currency=usd&days=30")
j_dog = res_dog.json()
df_dog = pd.DataFrame(j_dog["prices"], columns = ["time", "price"])
df_dog["time"] = pd.to_datetime(df_dog["time"], unit='ms')
df_dog.set_index('time', inplace=True)

fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(14,6))
axes[0].plot(df_lit)
axes[1].plot(df_dog)

#fig, ax = plt.subplots(figsize=(12,6))
#ax.plot(df_lit, color='b')
#ax.plot(df_dog, color='g')
#ax.set_ylim((-10,100))
```
